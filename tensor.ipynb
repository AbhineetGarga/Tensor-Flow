```python
import tensorflow as t
import matplotlib.pyplot as plt

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Conv2D
```

```python
print("TensorFlow version:", t.__version__)
```

```python
(train_images, train_labels), (test_images, test_labels) = t.keras.datasets.mnist.load_data()
```

```python
train_images = train_images / 255.0
test_images = test_images / 255.0
```

```python
print("Training set shape:", train_images.shape)
print("/n")
print("Training labels shape:", train_labels.shape)
print("/n")
print(train_labels)
```

```python
plt.imshow(train_images[0], cmap="gray")
plt.show()
```

```python
# my_model= t.keras.models.Sequential()
# my_model.add(t.keras.layers.Flatten(input_shape=(28,28)))
# my_model.add(t.keras.layers.Dense(128, activation="relu"))
# # t.keras.layers.Dropout(0.2),
# my_model.add(t.keras.layers.Dense(10, activation='softmax'))

my_model = t.keras.models.Sequential([
    t.keras.layers.Flatten(input_shape=(28, 28)),
    t.keras.layers.Dense(128, activation="relu"),
    t.keras.layers.Dense(10, activation="softmax")
])

```

```python
my_model.compile(optimizer='adam',
  loss='sparse_categorical_crossentropy',
  metrics=['accuracy'])
```

```python
my_model.fit(train_images, train_labels, epochs=100)
val_loss, val_acc= my_model.evaluate(test_images, test_labels)
print(f"Validation loss: {val_loss}")
print(f"Validation accuracy: {val_acc}")
```

```python
print("Training complete. Model evaluated on test data and the percentage accuracy percentage is:", val_acc*100)
```
